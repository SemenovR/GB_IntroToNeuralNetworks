### Решение:

Анализировал алгоритм EfficientDet по статье: https://ai.googleblog.com/2020/04/efficientdet-towards-scalable-and.html

Авторы статьи утвержают, что EfficientDet достигает точности уровня современных алгоритмов детектирования, будучи в 9 раз меньше и потребляя значительно меньше вычислений. В качестве базового алгоритма был выбран EfficientNet. Даже простая замена базового алгоритма ResNet-50 на EfficientNet-B3 позволила повысить точность на 3% при одновременном сокращении вычислений на 20%. Далее была заменена нисходящая пирамидальная сеть, на новую двунаправленную сеть (BiFPN) с улучшенной техникой быстрого нормализованного слияния и упрощённых алгоритмов свёртки. Благодаря чему точность повышается до 4%, снижая при этом затраты на вычисления на 50%. Третья оптимизация заключается в новом методе масштабирования, который совместно увеличивает разрешение / глубину / ширину. В частности каждая последующая модель EfficientDet от D1 до D7 обеспечивает более высокую точность. В результате при аналогичных условиях модели EfficientDet в 2–4 раза быстрее на GPU и в 5–11 раз быстрее на CPU, чем другие детекторы. Кроме того, немного изменив этот алгоритм можно получить алгоритм для семантической сегментации. Особых минусов у этого алгоритма не получилось обнаружить.
